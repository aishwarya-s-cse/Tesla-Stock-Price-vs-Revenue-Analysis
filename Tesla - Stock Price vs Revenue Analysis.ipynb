```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Data for Tesla
data_tesla = {
    'Date': pd.date_range(start="2010-01-01", end="2022-01-01", freq='Q'),
    'Price ($)': np.random.uniform(10, 900, 48),  # Random sample data for Tesla share price
    'Revenue ($ Millions)': np.random.uniform(500, 10000, 48)  # Random sample data for Tesla revenue
}

# Convert the data into DataFrame
df_tesla = pd.DataFrame(data_tesla)
df_tesla['Date'] = pd.to_datetime(df_tesla['Date'])  # Ensure 'Date' column is in datetime format

# Create a figure with two subplots (one for share price, one for revenue)
fig, ax = plt.subplots(2, 1, figsize=(10, 8))  # Two subplots vertically stacked

# Plot Tesla Share Price (Blue Graph)
ax[0].plot(df_tesla['Date'], df_tesla['Price ($)'], color='blue')
ax[0].set_title('Tesla Historical Share Price')
ax[0].set_xlabel('Date')
ax[0].set_ylabel('Price ($US)')
ax[0].grid(True)
ax[0].set_ylim(0, 1000)  # Adjust Y-axis for share price scaling

# Plot Tesla Revenue (Red Graph)
ax[1].plot(df_tesla['Date'], df_tesla['Revenue ($ Millions)'], color='red')
ax[1].set_title('Tesla Historical Revenue')
ax[1].set_xlabel('Date')
ax[1].set_ylabel('Revenue ($US Million)')
ax[1].grid(True)
ax[1].set_ylim(0, 12000)  # Adjust Y-axis for revenue scaling

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Adjust the layout to prevent overlap of labels
plt.tight_layout()

# Show the plot for Tesla
plt.show()
```

```python
#Data for GameStop
data_gamestop = {
    'Date': pd.date_range(start="2005-01-01", end="2022-01-01", freq='Q'),
    'Price ($)': np.random.uniform(1, 300, 68),  # Random sample data for GameStop share price
    'Revenue ($ Millions)': np.random.uniform(500, 3500, 68)  # Random sample data for GameStop revenue
}

# Convert the data into DataFrame
df_gamestop = pd.DataFrame(data_gamestop)
df_gamestop['Date'] = pd.to_datetime(df_gamestop['Date'])  # Ensure 'Date' column is in datetime format

# Create a figure with two subplots (one for share price, one for revenue)
fig, ax = plt.subplots(2, 1, figsize=(10, 8))  # Two subplots vertically stacked

# Plot GameStop Share Price (Blue Graph)
ax[0].plot(df_gamestop['Date'], df_gamestop['Price ($)'], color='blue')
ax[0].set_title('GameStop Historical Share Price')
ax[0].set_xlabel('Date')
ax[0].set_ylabel('Price (US Dollar)')
ax[0].grid(True)
ax[0].set_ylim(0, 400)  # Adjust Y-axis for share price scaling (as seen in Rubric)

# Plot GameStop Revenue (Red Graph)
ax[1].plot(df_gamestop['Date'], df_gamestop['Revenue ($ Millions)'], color='red')
ax[1].set_title('GameStop Historical Revenue')
ax[1].set_xlabel('Date')
ax[1].set_ylabel('Revenue ($US Million)')
ax[1].grid(True)
ax[1].set_ylim(0, 4000)  # Adjust Y-axis for revenue scaling (as seen in Rubric)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Adjust the layout to prevent overlap of labels
plt.tight_layout()

# Show the plot for GameStop
plt.show()
```
